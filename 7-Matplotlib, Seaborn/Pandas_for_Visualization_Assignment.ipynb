*`np.random.seed(1)` to get the repeatable results.*

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1)
df = pd.DataFrame(np.random.rand(1000,4),columns=['a', 'b', 'c','d'])
#df = pd.DataFrame(np.random.randn(1000,4),columns=['a', 'b', 'c','d'])
```

**You have created a dataset `df`. Plot `c vs d` as a scatter plot to recreate the plot below. Try to match the color and size of the data points. You should be able to match the size of the figure as well!**

```python
# Code goes here!
plt.figure(figsize=(12, 5))
plt.scatter(df['c'], df['d'], color='green')
plt.xlabel('c')
plt.ylabel('d')
plt.title('Scatter plot of c and d')
plt.show()
```

* Create a histogram of the 'd' column. *

```python
# Code goes here!
plt.hist(df['d'], bins=20)
plt.show()
```

** Change style sheets to set for 'bmh' and add more bins (e.g. 50) to the above plot.**

```python
# Code goes here!
plt.style.use('bmh')
plt.hist(df['d'], bins=50, color='blue')
plt.xlabel('d')
plt.ylabel('Frequency')
plt.title('Histogram of d with 50 bins')
plt.show()
```

**Create a scatter plot for a vs b. How to use the s parameter to indicate the size of the data point based of col d?**<br>

```python
# Code goes here!
plt.style.use('bmh')
plt.figure(figsize=(12, 5))
plt.scatter(df['a'], df['b'], s=df['d']*100, color='red')  # 's' parameter for size
plt.xlabel('a')
plt.ylabel('b')
plt.title('Scatter plot of a vs b with size based on d')
plt.show()
```

**Recreate the plot below**

```python
# Code goes here!
df.plot.hexbin(x='a', y='b', gridsize=30, cmap='coolwarm', figsize=(9, 3))
```

**Create a boxplot comparing all columns in the data.**<br>
Hint: use `color = dict(boxes='DarkGreen', whiskers='DarkOrange',medians='DarkBlue', caps='Gray')` and `sym = 'r+'` 

```python
# Code goes here!
boxprops = dict(color='DarkGreen')
whiskerprops = dict(color='DarkOrange')
capprops = dict(color='Gray')
medianprops = dict(color='DarkBlue')

df.boxplot(column=['a', 'b', 'c', 'd'], figsize=(8, 5), boxprops=boxprops, whiskerprops=whiskerprops, capprops=capprops, medianprops=medianprops, sym='r+')
plt.xlabel('Columns')
plt.ylabel('Values')
plt.title('Boxplot comparing all columns')
plt.show()
```
