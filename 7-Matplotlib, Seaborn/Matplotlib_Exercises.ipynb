# Plotting with Matplotlib

## Instructions:
1. **Use `plt.figure()` and create a figure object, say `fig`.**  
2. **Add an axis to the figure canvas (`fig`) using `add_axes()`.**  
   - The `[left, bottom, width, height]` is `[0, 0, 1, 1]`.  
3. **Plot the data `x, y` on the axes.**  
4. **Set the labels for x and y axes.**  
5. **Set the title of the plot.**  
6. **Create a legend as shown in the plot.**


```python
fig = plt.figure()
axis = fig.add_axes([0, 0, 1, 1])  # Left, bottom, width, height
plot1 = axis.plot(x, y, 'g')  # Set plot color to green
axis.set_xlabel('X Label')
axis.set_ylabel('Y Label')
axis.set_title('Plot Title')
legend1 = axis.legend(['y = x*2'], loc='lower right')  # loc=4
```

## Instructions  

1. **Create a figure object**  
2. **Put two axes on the figure at** `[0, 0, 1, 1]` **and** `[0.1, 0.6, 0.2, 0.2]`  
3. **Plot data (x, z) on the main axes**  
4. **Plot data (x, z) on the small axes and set its x-limit between 1 and 2, y-limit between 0 and 5**  
5. **Replicate the figure to include axes, titles, legends, etc.**  


```python
fig = plt.figure()
axis1 = fig.add_axes([0, 0, 1, 1])  # Left, bottom, width, height
axis2 = fig.add_axes([0.1, 0.6, 0.2, 0.2])  # Inset axes

# Plot data (x, z) on the main axes
plot1 = axis1.plot(x, z, 'r')  # Set plot color to red
axis1.set_xlabel('X')
axis1.set_ylabel('Z')
axis1.set_title('Main Plot')
axis1.legend(['Main Plot'], loc='upper left')

# Plot data (x, z) on the small axes
plot2 = axis2.plot(x, z, 'g')  # Set plot color to green
axis2.set_xlim(1, 2)
axis2.set_ylim(0, 5)
axis2.set_xlabel('X')
axis2.set_ylabel('Z')
axis2.set_title('Inset Plot')
axis2.legend(['Inset Plot'], loc='lower right', bbox_to_anchor=(1.2, 0))  # Move legend a little to the right

plt.show()
```

**3. Replicate the output using <code>subplots()</code> and for loop** -- Note, the data is x and y. 

```python
plt.subplot(1, 3, 1)
plt.plot(x, y, 'b-')
plt.subplot(1, 3, 2)
plt.plot(x, y, 'b-')
plt.subplot(1, 3, 3)
plt.plot(x, y, 'b-')

plt.tight_layout()
plt.show()
```

**4. Replicate the output using <code>subplots()</code>**

```python
# Create figure and axes
fig, axes = plt.subplots(3, 1, figsize=(10, 5))

# Adjust x and y ticks
for ax in axes:
    ax.set_xticks(np.arange(0, 1.1, 0.2))  # X-axis step of 0.2
    ax.set_yticks(np.arange(0, 1.1, 0.5))  # Y-axis step of 0.5

plt.tight_layout()
plt.show()
```

**5. plot `(x,y)`, `(x,z)` and `(x,y)` on the above empty canvas to replicate the plot given below.**

```python
fig, axes = plt.subplots(3, 1, figsize=(7, 5)) # figsize=(width, height)
plt.tight_layout()

# Plot data on each axes
axes[0].plot(x, y, 'g--')
axes[0].set_ylim(0, 10)  # Set y-axis limits for the first subplot

axes[1].plot(x, z, 'r')
axes[1].set_ylim(0, 50)  # Set y-axis limits for the second subplot

axes[2].plot(x, y, 'g')
axes[2].set_ylim(0, 10)  # Set y-axis limits for the third subplot

plt.show()
```

**6. Resize the above plots to `figsize = (3,4)`**

```python
#Code here
fig, axes = plt.subplots(3, 1, figsize=(3, 4))
plt.tight_layout()

# Plot data on each axes
axes[0].plot(x, y, 'g--')
axes[0].set_ylim(0, 10)  # Set y-axis limits for the first subplot

axes[1].plot(x, z, 'r')
axes[1].set_ylim(0, 50)  # Set y-axis limits for the second subplot

axes[2].plot(x, y, 'g')
axes[2].set_ylim(0, 10)  # Set y-axis limits for the third subplot

plt.show()
```
